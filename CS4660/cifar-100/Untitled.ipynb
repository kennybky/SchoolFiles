```python
import glob
import os
import cifar_predict as pred
import cv2

import matplotlib.pyplot as plt

%matplotlib inline

path = os.path.join('project_images', '*g')
files = glob.glob(path)
```

```python
results = []
for f in files:
    cl = pred.predict(f)
    results.append((f, cl))
```

```python
for item in results:
    f = item[0]
    cl = item[1]
    img = cv2.imread(f)
    plt.imshow(img)
    plt.show()
    print(cl)
```
