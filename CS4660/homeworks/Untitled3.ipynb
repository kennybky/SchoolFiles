### Created notebook. Import the CSV and read it


```python
import numpy as np
import pandas as pd
data = pd.read_csv('http://mlr.cs.umass.edu/ml/machine-learning-databases/adult/adult.data',
                   names = ["age","workclass","fnlwgt","education","education_num","marital-status","occupation",
                            "relationship","race","sex","capital-gain","capital-loss","hours-per-week",
                            "native-country","income-level"], sep=',\s', engine='python')
```

```python
data[0::20]
```

### Replace all empty numeric fields with the average of its column in this new column

```python
data.fillna(data.mean)
```

### for all values that are "?", find the value that is most common in the column it is found and replace the "?" with it.

```python
for column in data:
    data[column].replace("?", data[column].mode().iloc[0], inplace=True)
data[:30]
```

### Convert non-numeric columns (that is, categorical data) to columns with an integer representations (like we did with the decision trees).

```python
def encode_target(df_mod, target_column):
    targets = df_mod[target_column].unique()
    map_to_int = {name: n for n, name in enumerate(targets)}
    for name in map_to_int:
        df_mod[target_column].replace(name, map_to_int[name], inplace=True)

df = data.copy()

encode_target(df, "workclass")
encode_target(df, "education")
encode_target(df, "marital-status")
encode_target(df, "occupation")
encode_target(df, "relationship")
encode_target(df, "race")
encode_target(df, "sex")
encode_target(df, "native-country")
encode_target(df, "income-level")

df[:30]
```

### Convert the data to a numpy array if it isn't one already

```python
a = np.array(df)
a[0]
```

### Divide the data into a training and test set

```python
Y = a[:,-1]
X = a[:,0:-1]
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.3, random_state=2)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

### Run both a random forest and gaussian naive bayes on the training set

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score


rf_model =  RandomForestClassifier(max_depth=10,n_estimators=10)
rf_model.fit(X_train, Y_train)
```

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier(criterion="entropy")
clf.fit(X_train, Y_train)
```

### gaussian naive bayes 

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()

gnb.fit(X_train, Y_train)
```

### Compute the accuracy of predictions on the test set

### Random Forest

```python
acc_train = rf_model.score(X_train, Y_train)
acc_test = rf_model.score(X_test, Y_test)

print("Accuracy on training data:", acc_train)
print("Accuracy on test data:", acc_test)
```

### Gaussian Naive Bayes

```python
acc_train = gnb.score(X_train, Y_train)
acc_test = gnb.score(X_test, Y_test)

print("Accuracy on training data:", acc_train)
print("Accuracy on test data:", acc_test)
```
